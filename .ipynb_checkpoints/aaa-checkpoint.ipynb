```python
#Importamos las librerias necesarias

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import LabelEncoder
from sklearn import datasets, linear_model
from sklearn import model_selection
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import precision_score
from sklearn.metrics import accuracy_score, f1_score, recall_score
from sklearn.feature_selection import SelectKBest
import seaborn as sb
import matplotlib.pyplot as plt
import statsmodels.api as sm
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LogisticRegression
```

```python
#Leemos el archivo Csv 
df = pd.read_csv("weatherAus.csv")
#visualizamos las primeras 5 comlumnas del ddtaFrame
df.head()
```

```python
#visualizamos el tipo de los datos para ver si podemos transformar
df.dtypes
```

```python
#Transformamos las columnas "RainToday" y "RainTomorrow", 
#esto nos servira para que los modelos los podamos hacer sin ningun problema
le = LabelEncoder()
df['RainToday'] = le.fit_transform(df['RainToday' ])
#No=0 Y Si=1
df['RainTomorrow'] = le.fit_transform(df['RainTomorrow' ])
#No=0 Y Si=1
```

```python
#Visualizamos los tipos de datos, de esta manera verificamos que el tipo de dato se ha modificafo correctamente.
df.dtypes
```

```python
#realizamos un head para ver si la transformacion de los datos anteriores se realizo correctamente.
df.head()
```

```python
#Realizamos un describe de los datos,el que nos permitira ver los datos totales,
#el promedio,la media, la minima y maxima de los datos entre otros
df.describe()
```

```python
#visualizamos la cantidad de nulos en todas las columnas de nuestro data frame.
for feature in df.columns:
    print('Total de valores nulos de', feature, '=', df[feature].isna().sum())
```

```python
#Realizamos el tratamiento de los nulos y los transformamos en "0"
df.fillna(0, inplace = True)
```

```python
#Volvemos a visualizar si hay datos nulos,
#en este caso no hay ninguno, por lo que el tratamiento se ha realizado de forma correcta
for feature in df.columns:
    print('Total de valores nulos de', feature, '=', df[feature].isna().sum())
```

```python
#preguntamos si hay datos nulos una vez mas para asegurarnos.
df.isnull().any().any()
```

```python
%matplotlib inline

df.groupby('RainToday')['RainToday'].count().plot(kind='bar')
```

```python
#visualizamos el porcentaje de lluvia anual a lo largo de toda australia
porcentaje_lluvia = (df[df.Rainfall > 0]['Rainfall'].count() * 1.0/ df['Rainfall'].count()) * 100.0
print("El procentaje de lluvia en australia es {0:.2f}%"
.format(porcentaje_lluvia))
```

```python
#agrupamos los datos a utilizar para construir nuestro modelo
df1= df[["MinTemp","MaxTemp","Rainfall","Sunshine",
         "Cloud9am","Cloud3pm",
         "Temp9am","Temp3pm","RISK_MM"]]
        
```

```python
#visualizamos los datos elegidos
df1.head()
```

```python
#Definimos X,Y.
#Donde "X" son las columnas que utilizaremos e "Y" es nuestro target con el que nos ayudara a usar
#los modelos predictivos
X = df1
Y = df.RainTomorrow
```

```python
#Graficamos el porcentaje de lluvia el dia de maÃ±ana segun nuestro target "RainTomorrowu"
clases = np.array([df[df.RainTomorrow == 0.0].shape[0],
df[df.RainTomorrow == 1.0].shape[0]])
# Creando las leyendas del grafico.
labels = [ str(round(x * 1.0 / clases.sum() * 100.0, 2)) + '%' for x in clases]
labels[0] = 'No\n ' + labels[0]
labels[1] = 'Si\n ' + labels[1]
plt.pie(clases, labels=labels)
plt.title('Porcentaje de lluvia el dia de maÃ±ana')
plt.show()
```

```python
########################## MODELO NAIVE BAYES ##############################
########################## MODELO NAIVE BAYES ##############################
########################## MODELO NAIVE BAYES ##############################
```

```python
#Se separan los datos de "train" en entrenamiento y prueba para probar los algoritmos
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2,random_state =15)

#Definimos el algoritmo a utilizar que es Naive Bayes
modelo = GaussianNB()

#Entrenamiento del modelo
modelo.fit(X_train, Y_train)

#Validacion del modelo
Y_pred = modelo.predict(X_test)
```

```python
# EvaluaciÃ³n del modelo
matriz = confusion_matrix(Y_test, Y_pred)
print("Matriz de Confusion")
print(matriz)

#Se calcula la precision del modelo
precision = precision_score(Y_test, Y_pred)
print("Precision del modelo:", precision)
```

```python
#Usando otras medidas
accuracy = accuracy_score(Y_test, Y_pred)
print('Accuracy del modelo:',accuracy)
print('F1 score del modelo:',f1_score(Y_test, Y_pred))
print('Recall del modelo:',recall_score(Y_test, Y_pred))
```

```python
#Definimos variables
X = df1
y = df.RainTomorrow
# Separar los datos de "train" en entrenamiento y prueba para probar el modelo
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state = 29)
modelo = DecisionTreeRegressor(max_depth = 5, random_state = 29)
# Entrenamiento del modelo
modelo.fit(X_train, y_train)
# ValidaciÃ³n del modelo
Y_pred = modelo.predict(X_test)
# EvaluaciÃ³n del modelo
print('PrecisiÃ³n del modelo (en entrenamiento):', modelo.score(X_train,y_train))
print('PrecisiÃ³n del modelo (en validaciÃ³n):', modelo.score(X_test, y_test))
```

```python
#RepresentaciÃ³n grafica del arbol
from sklearn.tree import plot_tree
fig, ax = plt.subplots(figsize=(30, 10))
print(f"Profundidad del Ã¡rbol: {modelo.get_depth()}")
print(f"NÃºmero de nodos terminales: {modelo.get_n_leaves()}")
plot = plot_tree(
decision_tree = modelo,
feature_names = df1.columns,
    class_names = 'RainTomorrow',
    filled = True,
    impurity = False,
    fontsize = 10,
    precision = 2,
    ax = ax
)
```
